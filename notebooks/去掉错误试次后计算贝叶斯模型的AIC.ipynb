## 2. 尝试去掉超过3个标准差以外的数据再试一次

```r
result_list <- list()
sub_num_list <- unique(data_without_error$Subject_num)
count_num <- 1
for (i in sub_num_list) {
    tmp_data <- filter(data_without_error, Subject_num == i)
    if_outlier_list <- vector(length = nrow(tmp_data), mode = "numeric")
    mean_rt <- mean(tmp_data$RT)
    sd_rt <- sd(tmp_data$RT)
    for (j in 1:nrow(tmp_data)) {
        if (tmp_data$RT[j] > mean_rt + 3 * sd_rt ||
            tmp_data$RT[j] < mean_rt - 3 * sd_rt) {
            if_outlier_list[j] <- 1
        } else {
            if_outlier_list[j] <- 0
        }
    }
    result_list[[count_num]] <- cbind(tmp_data, if_outlier = if_outlier_list)
    count_num <- count_num + 1
}
```

```r
tagged_data <- bind_rows(result_list, .id = "column_label")
```

```r
data_without_outlier <- filter(tagged_data, if_outlier == 0)
```

```r
prep_lm_data_no_outlier <- dummy_columns(data_without_outlier, "miniblock") %>%
                           select(RT, Subject_num, starts_with("miniblock_"),
                           ab_r_selected, sr_r_selected)
```

## 4. 尝试使用随机效应模型理解变量对 RT 的解释

```r
library(lme4)
library(lmerTest)
```

```r
data_without_error <- filter(all_sub_data, Type == "hit")
model1 <- lmer(RT ~ ab_r_selected + (ab_r_selected | Subject_num) + (ab_r_selected | run), data = data_without_error)
summary(model1)
```

```r
data_without_error <- filter(all_sub_data, Type == "hit")
model2 <- lmer(RT ~ sr_r_selected + (sr_r_selected | Subject_num) + (sr_r_selected | run), data = data_without_error)
summary(model2)
```

```r
anova(model1, model2)
```

## 5. 对强化学习模型进行 AIC 估计

```r
all_sub_data_rl <- read_csv("/Users/dddd1007/project2git/cognitive_control_model/data/input/all_data_with_rl_model.csv") # nolint
rl_data_without_error <- filter(all_sub_data_rl, Type == "hit")
sub_num_list <- unique(rl_data_without_error$Subject_num)
count_num <- 1
result_list <- list()
for (i in sub_num_list) {
    tmp_data <- filter(rl_data_without_error, Subject_num == i)
    if_outlier_list <- vector(length = nrow(tmp_data), mode = "numeric")
    mean_rt <- mean(tmp_data$RT)
    sd_rt <- sd(tmp_data$RT)
    for (j in 1:nrow(tmp_data)) {
        if (tmp_data$RT[j] > mean_rt + 3 * sd_rt ||
            tmp_data$RT[j] < mean_rt - 3 * sd_rt) {
            if_outlier_list[j] <- 1
        } else {
            if_outlier_list[j] <- 0
        }
    }
    result_list[[count_num]] <- cbind(tmp_data, if_outlier = if_outlier_list)
    count_num <- count_num + 1
}
tagged_data <- bind_rows(result_list, .id = "column_label")
rl_data_without_outlier <- filter(tagged_data, if_outlier == 0)
rl_data_without_outlier <- unite(rl_data_without_error, "miniblock",
                                 block, run, sep = "_")
lm_result_ab <- dummy_columns(rl_data_without_outlier, "miniblock") %>%
                    select(RT, Subject_num, starts_with("miniblock_"), AB) %>%
                    split(.$Subject_num) %>%
                    map(select, -Subject_num) %>%
                    map( ~lm(RT ~ ., data = .)) %>%
                    map(AIC)
lm_result_sr <- dummy_columns(rl_data_without_outlier, "miniblock") %>%
                    select(RT, Subject_num, starts_with("miniblock_"), SR) %>%
                    split(.$Subject_num) %>%
                    map(select, -Subject_num) %>%
                    map( ~lm(RT ~ ., data = .)) %>%
                    map(AIC)
lm_result_sr_decay <- dummy_columns(rl_data_without_outlier, "miniblock") %>%
                    select(RT, Subject_num, starts_with("miniblock_"), SR_Decay) %>%
                    split(.$Subject_num) %>%
                    map(select, -Subject_num) %>%
                    map( ~lm(RT ~ ., data = .)) %>%
                    map(AIC)
```
